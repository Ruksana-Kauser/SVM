```python
#import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading dataset

ff_data = pd.read_csv('forestfires.csv')
ff_data
```

```python
#data understanding

ff_data.info()
```

```python
sns.distplot(ff_data['size_category'])
plt.show()
```

```python
sns.boxplot(data=ff_data)
plt.show()
```

```python
#data preprocessing

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
ff_data['month']=le.fit_transform(ff_data['month'])
ff_data['day']=le.fit_transform(ff_data['day'])
ff_data['size_category']=le.fit_transform(ff_data['size_category'])
```

```python
ff_data.info()
```

```python
ff_data.head()
```

```python
#model building

x=ff_data.iloc[:,0:30]
y=ff_data.iloc[:,30]
```

```python
x
```

```python
y
```

```python
# splitting data into train and test

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=12)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
#SVM Classification using Kernels: linear,poly,rbf

from sklearn.svm import SVC 
```

```python
#kernel = linear
linear_model = SVC(kernel='linear')
linear_model.fit(x_train,y_train)
train_linear_pred = linear_model.predict(x_train)
test_linear_pred = linear_model.predict(x_test)
train_linear_acc = np.mean(train_linear_pred==y_train)
test_linear_acc = np.mean(test_linear_pred==y_test)
```

```python
train_linear_acc
```

```python
test_linear_acc
```

```python
#kernel=poly
poly_model=SVC(kernel='poly')
poly_model.fit(x_train,y_train)
train_poly_pred=poly_model.predict(x_train)
test_poly_pred=poly_model.predict(x_test)
train_poly_acc=np.mean(train_poly_pred==y_train)
test_poly_acc=np.mean(test_poly_pred==y_test)
```

```python
train_poly_acc
```

```python
test_poly_acc
```

```python
#kernel = rbf
rbf_model=SVC(kernel='rbf')
rbf_model.fit(x_train,y_train)
train_rbf_pred=rbf_model.predict(x_train)
test_rbf_pred=rbf_model.predict(x_test)
train_rbf_acc=np.mean(train_rbf_pred==y_train)
test_rbf_acc=np.mean(test_rbf_pred==y_test)
```

```python
train_rbf_acc
```
